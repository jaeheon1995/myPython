##### 그래프의 종류
**막대그래프** : 데이터를 카테고리별로 값을 막대형태로 비교하고자 할 때 사용하는 그래프 형식
(수직,수평  막대그래프)
>카테고리별 데이터를 비교하고자 할 때 사용(제품 판매량,지역별 인구수)

**선 그래프** : 시간의 흐름에 따라 데이터 포인터를 선으로 연결하여 사용하는 그래프 형식
>하나의 요소에 대해 시간의 흐름에 따라 변경값을 나타내고자 할 때(기온,주식)

**파이차트** : 전체 데이터에 대한 각 요소의 비율을 원형으로 나타 낼 때 사용하는 그래프 형식
>전체에 대한 요소의 비율을 표현하고자 할 때 사용(투표 결과)

**산전도** : 두 변수 간의 상관관계를 표현할 때 사용하는 그래프 형식
> 각 요소간 관계성을 파악할 때 사용(키와 몸무게 상관관계 비교)

**히스토그램** : 범위별로 그룹화하여 각 그룹에 속한 데이터의 수를 막대로 표현하는 그래프형식
>데이터의 분포를 비교할 때 사용(시험 점수)

**영역 그래프** : 선그래프의 아래쪽 빈공간을 채운그래프
> 시간의 흐름에따라 데이터를 비교할 때 사용 변화량을 강조하고자 할때 사용

**박스 그래프** : 데이터의 분포를 나타낼때 사용하는 그래프형식
>최대값 최소값 사분위수로 데이터의 이상치를 판단할때 사용

### 선그래프
가장 기본적인 그래프 형식 시간의 흐름 및 순서에 따라서 데이터의 변화량을 보여주고자 할때 사용
plot패키지의 plot메서드로 간단하게 생성가능

````python
plt.title('plot')
plt.plot([1,2,4,8,16])
plt.show
````

````python
#그래프의 제목을 지정할 때는 .title()메서드를 사용
#plot(x 축에대한 리스트,y축에 대한 리스트)
plt.title('x 축 추가')  
plt.plot([2021,2022,2023,2024,2025], [3,5,7,9,21])
plt.show
````

````python
# 차트 스타일 지정
# plot()메서드의 세번째 매개변수로 지정된 문자열 포멧으로 전달하여 스타일 지정가능
plt.title('c2-- 스타일 적용')
plt.plot([2021,2022,2023,2024,2025], [3,5,7,9,21] , 'c2--')
plt.show
````

````python
#스타일 매개변수
#color (c) : 선색상,linewidth(lw):선굵기 , linestyle(ls): 선스타일
#marker: 마커종류, markersize(ms) : 마커크기 markeredgecoler (mec):마커테두리 색상 
#markeredgewid(mew) : 마커테두리굵기 markerfacecoler(mfc): 마커내부색상
plt.title('c2-- 스타일 적용')
plt.plot([2021,2022,2023,2024,2025], [3,5,7,9,21] , c='#00998850',lw = 5, ls=':',marker='*', ms=10,mfc='#009988')
plt.show
````

````python
#차트범위지정
#xlim(),ylim() 메서드로 x,y축의 상,하한을 지정
plt.title('차트범위지정')
plt.plot([2021,2022,2023,2024,2025], [3,5,7,9,21] , c='#00998850',lw = 5, ls=':',marker='*', ms=10,mfc='#009988')
plt.xlim(2019,2025)
plt.ylim(3,20)
plt.show
````

````python
#틱 지정
#xtikes(),ytikes() 메서드로 틱의 위치나 라벨 지정 가능
plt.title('틱지정')
plt.plot([2020,2022,2023,2024,2025], [3,5,7,9,21] , c='#00998850',lw = 5, ls=':',marker='*', ms=10,mfc='#009988')
plt.xticks([2020,2022,2024])
plt.yticks([0,8,16],['저점','중간점','고점'])
plt.show
````

````python
#그리드지정
#grid()메서드를 통해서 그리드 표시여부 결정가능
plt.title('그리드지정')
plt.plot([2020,2022,2023,2024,2025], [3,5,7,9,21] , c='#00998850',lw = 5, ls=':',marker='*', ms=10,mfc='#009988')
plt.xticks([2020,2022,2024])
plt.yticks([0,8,16],['저점','중간점','고점'])
plt.grid(True)
plt.show
````

````python
#차트겹치기
#plot()메서드를 중복해서호출하여 차트를 겹칠수있다

plt.title('겹쳐그리기')
plt.plot([2020,2022,2023,2024,2025], [3,5,7,9,21] , c='#00998850',lw = 5, ls=':',marker='*', ms=10,mfc='#009988')
plt.plot([2020,2022,2023,2024,2025], [16,8,4,2,1] , c='#88990050',lw = 5, ls=':',marker='*', ms=10,mfc='#889900')
plt.xticks([2020,2022,2024])
plt.yticks([0,8,16],['저점','중간점','고점'])
plt.grid(True)
plt.show
````

````python
#범례
#legend()메서드를 통해 범례를 표시
#loc매개변수에 위치에 해당하는 정수를 전달하여 위치 설정 가능
plt.title('범례')
plt.plot([2020,2022,2023,2024,2025], [3,5,7,9,21] , c='#00998850',lw = 5, ls=':',marker='*', ms=10,mfc='#009988', label = '증가')
plt.plot([2020,2022,2023,2024,2025], [16,8,4,2,1] , c='#88990050',lw = 5, ls=':',marker='*', ms=10,mfc='#889900', label ='감소')
plt.xticks([2020,2022,2024])
plt.yticks([0,8,16],['저점','중간점','고점'])
plt.grid(True)
plt.legend(loc=0)
plt.show
````

````python
#각 틱의 라벨
#xlabel(),ylabel()메서드로 각축의 라벨표시 가능

plt.title('각 틱의 라벨')
plt.plot([2020,2022,2023,2024,2025], [3,5,7,9,21] , 
         c='#00998850',lw = 5, ls=':',marker='*', ms=10,mfc='#009988', label = '증가')
plt.plot([2020,2022,2023,2024,2025], [16,8,4,2,1] , 
         c='#88990050',lw = 5, ls=':',marker='*', ms=10,mfc='#889900', label ='감소')
plt.xticks([2020,2022,2024])
plt.yticks([0,8,16],['저점','중간점','고점'])
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('연도')
plt.ylabel('가치')
plt.show
````

### 막대그래프
x축의 값이 분류가 가능한 카테고리 형태일때 bar(),barh() 메서드로 막대그래프를 표현가능  


````python
plt.title('막대그래프')
plt.bar(['a','b','c'],[20,78,43], color ='#998832')
plt.show
````

````python
#가로막대그래프 barh()메서드로 생성가능
#오차표시
#xerr,yerr 매개변수이용
plt.title('막대그래프')
plt.barh(['a','b','c'],[20,78,43],xerr=[2,10,4], color ='#365250')
plt.show
````

### 스템 플롯
기준점을 중심으로 오차범위를 표현하고자 할때 사용하는 차트

````python

x = np.linspace(0.1, 2 * np.pi, 10)
plt.title('스템플롯')
plt.stem(x, np.cos(x))
plt.show()
````

### 파이차트
각 분류가 전체에서 차지하는 비중을 표현할 때 사용하는 차트  


````python
plt.title('파이차트')
plt.axis('equal')
plt.pie([25, 45, 30], labels=['사과','바나나','포도'],
        explode=[0,0.1,0], colors=['#ff000050','#00ff22','#232366']
        ,shadow=True, autopct='%1.1f%%')
plt.show
````

### 컨투어플롯
3차원 자료에서 x,y 입력과 z출력이 존재할 때,z를 선과 색상으로 표현하는방법


````python
def func(x,y):
    z = (1 - x / 2 + x ** 5 + y **3 )* np.exp(-x ** 2 - y **2)

    return z

x= np.linspace(-3,3,256)
y= np.linspace(-3,3,256)
xx,yy = np.meshgrid(x,y)
zz = func(xx,yy)

plt.title('등고선 그래프')
plt.contour(xx,yy,zz, colors ='black')
plt.contourf(xx,yy,zz)
plt.grid(True)
plt.show
````
